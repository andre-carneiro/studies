Treating duplicates with .drop_duplicates() method
- subset : List of column names to check for duplication
- keep : Whether to keep first ("first"), last ("last") or all (False) duplicate values
- inplace: Drop duplicated rows directly inside DataFrame without creating new object (True)

```python
# Column names to check for duplication
column_names = ['first_name','last_name','address'] #check for all information
duplicates = height_weight.duplicated(subset = column_names, keep = False) #keep all duplicates

# DROP DUPLICATES
height_weight.drop_duplicates(inplace = True)

# Output duplicate values
height_weight[duplicates]
height_weight[duplicates].sort_values(by = 'first_name') #verify all of the duplicates with sort method
```

```python
#TREAT DUPLICATES (groupby and agg)
# Group by column names and produce statistical summaries
column_names = ['first_name','last_name','address']
summaries = {'height': 'max', 'weight': 'mean'}
height_weight = height_weight.groupby(by = column_names).agg(summaries).reset_index() #reset the index and use the default one instead

# Make sure aggregation is done
duplicates = height_weight.duplicated(subset = column_names, keep = False)
height_weight[duplicates].sort_values(by = 'first_name')
```

## Does Missingness Have A Pattern?

### Is the data missing at random?

Possible reasons for missing data
- Note â (variable â data field or column in a DataFrame)
    - Values simply missing at random instances or intervals in a variable
    - Values missing due to another variable
    - Values missing due to the missingness of the same or another variable

Types of missingness
1. Missing Completely at Random (MCAR)
2. Missing at Random (MAR)
3. Missing Not at Random (MNAR)

Missing Completely at Random(MCAR)
- Definition: "Missingness has no relationship between any values, observed or missing"


MCAR - An example

```python
msno.matrix(diabetes) #no correlation between variables missingness
```

Missing at Random(MAR)
- Definition: "There is a systematic relationship between missingness and other observed data, but not the missing data"
- Students sitting closer might have a fever

MAR - An example

```python
msno.matrix(diabetes)
```

Missing not at Random (MNAR)

- Definition: "There is a relationship between missingness and its values, missing or non-missing"
- (birthday party)


MNAR - An example
- Missingness pattern of the diabetes sorted by Serum_Insulin

```python
sorted = diabetes.sort_values('Serum_Insulin')
msno.matrix(sorted)
```

Summary
- Possible reasons for missingness
    - Missing Completely at Random (MCAR),
    - Missing at Random (MAR) or
    - Missing Not at Random (MNAR)
- Detecting missingness pattern by sorting the variables (and visualizing)
- Mapping missingness to MCAR, MAR & MNAR
