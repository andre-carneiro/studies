### Plotting time-series with different variables

```python
#Plotting two time-series together
import pandas as pd
climate_change = pd.read_csv('climate_change.csv',
                             parse_dates=["date"],
                             index_col="date")
climate_change
```

```python
#Plotting two time-series together
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax.plot(climate_change.index, climate_change["co2"])
ax.plot(climate_change.index, climate_change["relative_temp"])
ax.set_xlabel('Time')
ax.set_ylabel('CO2 (ppm) / Relative temperature')
plt.show()
```

```python
#Using twin axes (method) and separaing variables by color
fig, ax = plt.subplots()

ax.plot(climate_change.index, 
        climate_change["co2"], 
        color='blue')
ax.set_xlabel('Time')
ax.set_ylabel('CO2 (ppm)', 
              color='blue')

ax2 = ax.twinx() #share the same x, but not the same y

ax2.plot(climate_change.index, 
         climate_change["relative_temp"],
         color='red')
ax2.set_ylabel('Relative temperature (Celsius)', 
               color='red')
plt.show()
```

```python
#Coloring the ticks
fig, ax = plt.subplots()

ax.plot(climate_change.index, 
        climate_change["co2"],
        color='blue')
ax.set_xlabel('Time')
ax.set_ylabel('CO2 (ppm)', 
              color='blue')
ax.tick_params('y', 
               colors='blue')

ax2 = ax.twinx()

ax2.plot(climate_change.index,
         climate_change["relative_temp"],
         color='red')
ax2.set_ylabel('Relative temperature (Celsius)', 
               color='red')
ax2.tick_params('y', 
                colors='red')

plt.show()
```

```python
#A function that plots time-series
def plot_timeseries(axes, x, y, color, xlabel, ylabel): # Define a function called plot_timeseries
  axes.plot(x, y, color=color)   # Plot the inputs x,y in the provided color
  axes.set_xlabel(xlabel)   # Set the x-axis label
  axes.set_ylabel(ylabel, color=color)  # Set the y-axis label
  axes.tick_params('y', colors=color) # Set the colors tick params for y-axis

#Using our function
fig, ax = plt.subplots()

plot_timeseries(ax, climate_change.index, climate_change['co2'],
                'blue', 'Time', 'CO2 (ppm)')

ax2 = ax.twinx()

plot_timeseries(ax2, climate_change.index, climate_change['relative_temp'], 
                'red', 'Time', 'Relative temperature (Celsius)')

plt.show()
```

### Annotating time-series data

Annotations focus on specific points of the data and help to explain and interpret its features: https://matplotlib.org/2.0.2/users/annotations.html

```python
#Annotation
fig, ax = plt.subplots()
plot_timeseries(ax, climate_change.index, climate_change['co2'],
                'blue', 'Time', 'CO2 (ppm)')
ax2 = ax.twinx()
plot_timeseries(ax2, climate_change.index,
                climate_change['relative_temp'],
                'red', 'Time', 'Relative temperature (Celsius)')
ax2.annotate(">1 degree", 
             xy=[pd.Timestamp("2015-10-06"), 1]) #string larger than 1 degree and xy coordinate to annotate. TimeStamp object helps and y is 1 degree celsius of the data
plt.show()
```

```python
#Positioning the text
fig, ax = plt.subplots()
plot_timeseries(ax, climate_change.index, climate_change['co2'],
                'blue', 'Time', 'CO2 (ppm)')
ax2 = ax.twinx()
plot_timeseries(ax2, climate_change.index,
                climate_change['relative_temp'],
                'red', 'Time', 'Relative temperature (Celsius)')
ax2.annotate(">1 degree",
             xy=(pd.Timestamp('2015-10-06'), 1),
             xytext=(pd.Timestamp('2008-10-06'), -0.2)) #select xy position of the text. However, can not see where it is being annotated
```

```python
#Adding arrows to annotation
fig, ax = plt.subplots()
plot_timeseries(ax, climate_change.index, climate_change['co2'],
                'blue', 'Time', 'CO2 (ppm)')
ax2 = ax.twinx()
plot_timeseries(ax2, climate_change.index,
                climate_change['relative_temp'],
                'red', 'Time', 'Relative temperature (Celsius)')
ax2.annotate(">1 degree",
             xy=(pd.Timestamp('2015-10-06'), 1),
             xytext=(pd.Timestamp('2008-10-06'), -0.2),
             arrowprops={}) #arrow that indicates the point and properties of arrows
```

```python
#Customizing arrow properties
fig, ax = plt.subplots()
plot_timeseries(ax, climate_change.index, climate_change['co2'],
                'blue', 'Time', 'CO2 (ppm)')
ax2 = ax.twinx()
plot_timeseries(ax2, climate_change.index,
                climate_change['relative_temp'],
                'red', 'Time', 'Relative temperature (Celsius)')
ax2.annotate(">1 degree",
             xy=(pd.Timestamp('2015-10-06'), 1),
             xytext=(pd.Timestamp('2008-10-06'), -0.2),
             arrowprops={"arrowstyle":"->", "color":"gray"})

#xytext is the location
```

```python
from google.colab import files
uploaded = files.upload()

import io
import pandas as pd

climate_change = pd.read_csv(io.StringIO(uploaded['climate_change.csv'].decode('utf-8')), sep = ",")
print(climate_change.head())

#transform the date column into index
climate_change["date"] = pd.to_datetime(climate_change["date"])
climate_change = climate_change.set_index("date").sort_index()

print(climate_change.head())
print(climate_change.info())
```

```python
#Introducing scatter plots
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
ax.scatter(climate_change["co2"], climate_change["relative_temp"])
ax.set_xlabel("CO2 (ppm)")
ax.set_ylabel("Relative temperature (Celsius)")
plt.show()
```

```python
#Customizing scatter plots
eighties = climate_change["1980-01-01":"1989-12-31"]
nineties = climate_change["1990-01-01":"1999-12-31"]
fig, ax = plt.subplots()

ax.scatter(eighties["co2"], eighties["relative_temp"], 
           color="red", label="eighties")
ax.scatter(nineties["co2"], nineties["relative_temp"], 
           color="blue", label="nineties")

ax.legend()
ax.set_xlabel("CO2 (ppm)")
ax.set_ylabel("Relative temperature (Celsius)")
plt.show()
```

```python
#Encoding a third variable by color
fig, ax = plt.subplots()
ax.scatter(climate_change["co2"], climate_change["relative_temp"], 
           c=climate_change.index) #different from color argument
           
ax.set_xlabel("CO2 (ppm)")
ax.set_ylabel("Relative temperature (Celsius)")
plt.show()
```

General guidelines for choosing plotting style:
- Dark backgrounds are usually less visible (do not use it)
- If color is important, considter choosing colorblind-friendly options
  - "seaborn-colorblind" or
  - "tableau-colorblind10"
- If you think that someone will want to print your figure, use less ink
- If it will be printed in black-and-white, use the "grascale" style
