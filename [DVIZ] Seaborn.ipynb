## Introduction to Seaborn

## Visualizing Two Quantitative Variables

### Introduction to relational plots and subplots

- Scatterplots helps to identify relationship between quantitative variables
- Scatterplots may be different depending on the category that we are analyzing
- relplot() (instead of scatterplot())
  - relplot() lets you create subplots in a single figure

```python
#Using scatterplot()
import seaborn as sns
import matplotlib.pyplot as plt
sns.scatterplot(x="total_bill", y="tip",data=tips)
plt.show()

#Using relplot() - helps with scatter and line plots
import seaborn as sns
import matplotlib.pyplot as plt
sns.relplot(x="total_bill",y="tip",data=tips,kind="scatter") #kind parameter is the new one
plt.show()
```

```python
#Subplots in columns
import seaborn as sns
import matplotlib.pyplot as plt
sns.relplot(x="total_bill",y="tip",data=tips,
            kind="scatter",col="smoker") #argument col smoker separate the dataset smoker in columns
plt.show()
```

```python
#Subplots in rows
import seaborn as sns
import matplotlib.pyplot as plt
sns.relplot(x="total_bill",y="tip",data=tips,
            kind="scatter",row="smoker") #argument row smoker separate the dataset smoker in rows
plt.show()
```

- Changing the whiskers using `whis`
- By default, the whiskers extend to 1.5 * the interquartile range
- Make them extend to 2.0 * IQR: whis=2.0
- Show the 5th and 95th percentiles: whis=[5, 95]
- Show min and max values: whis=[0, 100]

### Point plots

- Show the of mean of quantitative variable for observations in each category
- Vertical lines show 95% confidence intervals (and level of uncertainty) - we assume that it is a sample data of a population

- Difference of pont plot and line plot
- Both show mean of quantitative variable with 95%  of confidence intervals
  - Line plots are relational plots. Has quantitative variable (usually time) on x-axis
  - Point plot has categorical varible on x-axis (easier to compare instead of high of the bars)
