```python
#Example
# Print the list created using the Non-Pythonic approach
i = 0
new_list= []
while i < len(names):
    if len(names[i]) >= 6:
        new_list.append(names[i])
    i += 1
print(new_list)

# Print the list created by looping over the contents of names
better_list = []
for name in names:
    if len(name) >= 6:
        better_list.append(name)
print(better_list)

# Print the list created by using list comprehension
best_list = [name for name in names if len(name) >= 6]
print(best_list)
```

NumPy array broadcasting
- Python lists don't support broadcasting

```python
nums = [-2,-1, 0, 1, 2]
nums ** 2
#output: TypeError: unsupported operand type(s) for ** or pow(): 'list' and 'int'
```

List approach (even with list comprehension, not the best one)

```python
# For loop (inefficient option)
sqrd_nums = []
for num in nums:
    sqrd_nums.append(num ** 2)
print(sqrd_nums)
#output: [4, 1, 0, 1, 4]

# List comprehension (better option but not best)
sqrd_nums = [num ** 2 for num in nums]
print(sqrd_nums)
#output: [4, 1, 0, 1, 4]
```

NumPy array broadcasting
- NumPy array broadcasting for the win!
- Vectorize operations. Very efficient

```python
nums_np = np.array([-2,-1, 0, 1, 2])
nums_np ** 2
#output: array([4, 1, 0, 1, 4])
```

### Examining runtime

Why should we time our code?
- Allows us to pick the optimal coding approach
- Faster code == more efficient code!

How can we time our code?
- Calculate runtime with IPython magic command %timeit
- Magic commands: enhancements on top of normal Python syntax
    - Prefixed by the "%" character
    - Link to docs (here - https://ipython.readthedocs.io/en/stable/interactive/magics.html)
    - See all available magic commands with %lsmagic

Using %timeit
- Code to be timed

```python
import numpy as np
rand_nums = np.random.rand(1000) #0 to 1
```

Timing with %timeit

```python
%timeit rand_nums = np.random.rand(1000)
#output: 8.61 µs ± 69.1 ns per loop (mean ± std. dev. of 7 runs, 100000 loops each)
```

- mean, std dev, makes several ones

Specifying number of runs/loops
- Seing the number of runs ( -r ) and/or loops ( -n )

```python
# Set number of runs to 2 (-r2)
# Set number of loops to 10 (-n10)
%timeit -r2 -n10 rand_nums = np.random.rand(1000)

#output: 16.9 µs ± 5.14 µs per loop (mean ± std. dev. of 2 runs, 10 loops each)
```

Using %timeit in line magic mode
- Line magic ( %timeit )

```python
# Single line of code
%timeit nums = [x for x in range(10)]
#output: 914 ns ± 7.33 ns per loop (mean ± std. dev. of 7 runs, 1000000 loops each)
```

Using %timeit in cell magic mode
- Cell magic ( %%timeit )

```python
# Multiple lines of code
%%timeit
nums = []
for x in range(10):
    nums.append(x)
#output: 1.17 µs ± 3.26 ns per loop (mean ± std. dev. of 7 runs, 1000000 loops each)
```

Saving output
- Saving the output to a variable ( -o )

```python
times = %timeit -o rand_nums = np.random.rand(1000)
#output: 8.69 µs ± 91.4 ns per loop (mean ± std. dev. of 7 runs, 100000 loops each)
```

times.timings

```python
heroes = ['Batman','Superman','Wonder Woman']
hts = np.array([188.0, 191.0, 183.0])
wts = np.array([ 95.0, 101.0, 74.0])
```

```python
def convert_units(heroes, heights, weights):
    new_hts = [ht * 0.39370 for ht in heights] #cm -> inches
    new_wts = [wt * 2.20462 for wt in weights] #kg -> pounds
    hero_data = {}
    
    for i,hero in enumerate(heroes):
        hero_data[hero] = (new_hts[i], new_wts[i])
        
    return hero_data
```
